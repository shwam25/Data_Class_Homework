```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sqlalchemy import create_engine
import psycopg2
import databaseconfig as cfg
```

```python
#Connection to Postgres database
engine = create_engine(f'postgresql://{cfg.mysql["user"]}:{cfg.mysql["password"]}@{cfg.mysql["host"]}:5432/{cfg.mysql["db"]}')
conn = engine.connect()
```

```python
#Create DataFrames from employees, salaries, titles tables in the Database
employees_df = pd.read_sql_table('employees', engine)
salaries_df = pd.read_sql_table('salaries', engine)
titles_df = pd.read_sql_table('titles', engine)
```

```python
#Create a histogram to visualize the most common salary ranges for employees
# Set Range
x_bins = 8

# Create Histogram of Number of Employees for each salary range
fig, ax = plt.subplots(figsize=(10,20))
x,bins,c = plt.hist(salaries_df["salary"], bins=x_bins,edgecolor='grey')

# Set the X Axis Ticks to be the Start/End of each range, display vertically
ax.set_xticks(bins)
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)

# Get the actual number for each range and display at the top of the bar
for i in range(x_bins):
    emp_count = int(x[i])
    plt.annotate(str(emp_count),(bins[i]+2000,x[i]+1000),fontsize=20)

plt.ylabel("Number of Employees",fontsize=10)
plt.xlabel("Salary",fontsize=10)
plt.title("Salary Ranges for Employees",fontsize=10)
plt.savefig("BarChart and Histogram PNGs\Histogram_Salary Ranges for Employees")
plt.show()
```

```python
# Create a query and DataFrame to store salary by titles
query = "SELECT title,avg(salary) as salary FROM salaries JOIN employees on salaries.emp_no=employees.emp_no JOIN titles on titles.title_id=employees.emp_title_id GROUP BY title ORDER BY salary DESC;"
Salary_by_titles = pd.read_sql_query(query,conn)
```

```python
# Create Bar Chart of Average Salary by Title
plt.bar(Salary_by_titles["title"],Salary_by_titles["salary"],edgecolor="grey")
plt.xticks(rotation=90)
plt.xlabel("Titles")
plt.ylabel("Salary")
plt.title("Average Salary by Title")
plt.savefig("BarChart and Histogram PNGs\Bar Chart_Average Salary by Title")
plt.show()
```
